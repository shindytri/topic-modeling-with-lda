```python
import gensim
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_csv('data/preprocessed_data.csv')
#Tokenization
# df['Tokenization'] = 
df.Preprocessing = df.Preprocessing.str.split()
df.head()
```

```python
# create a dictionary, to count how many time a word appears in dataset
dictionary = gensim.corpora.Dictionary(df.Preprocessing)

# corpus containing information about what words it contain and how many time it appears per document/Tweet
corpus = [dictionary.doc2bow(text) for text in df.Preprocessing]

# dictionary.filter_extremes(no_below=5, no_above=5, keep_n=100000)
print(len(corpus),len(dictionary))
```

```python
idx = 1
for x in corpus:
    print("Corpus "+str(idx))
    for i in range(len(x)):
        print(dictionary[x[i][0]],':',x[i][1])
    idx+=1
    print()
```

```python
tfidf = gensim.models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]

for doc in corpus_tfidf:
    pprint(doc)
    break
```

```python
lda_model = gensim.models.LdaMulticore(corpus, id2word=dictionary, num_topics=10)
for topic in lda_model.print_topics(-1):
    print(topic)
    print()
```

```python
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, id2word=dictionary, num_topics=10)
for topic in lda_model_tfidf.print_topics(-1):
    print(topic)
    print()
```

```python
for index, score in sorted(lda_model[corpus[1234]], key=lambda tup: -1*tup[1]):
    print("Score:",score)
    print("Topic:",lda_model.print_topic(index, 10),"\n")
```
